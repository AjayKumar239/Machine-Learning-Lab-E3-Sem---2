```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
x_axis = np.array([
    1.2, 2.2, 2.0, 1.7, 2.3,       # Cluster 1
    7.8, 8.2, 8.0, 7.9, 8.3,       # Cluster 2
    4.8, 5.1, 5.0, 4.9, 5.2 
      ])
y_axis = np.array([
     2.2, 1.9, 2.1, 2.3, 1.8,       # Cluster 1
    3.2, 3.1, 2.8, 3.0, 3.3,       # Cluster 2
    8.2, 8.3, 7.9, 8.1, 7.8    

])


data = pd.DataFrame({'x': x_axis , 'y': y_axis})

plt.figure(figsize=(8 ,8))

plt.plot()

plt.scatter(x_axis , y_axis)

plt.title("Cluster ")
plt.grid()

plt.show()
```

```python
Kmeans = KMeans(n_clusters= 1 , max_iter= 100)
Kmeans.fit(data)
```

```python
data['clusters'] = Kmeans.labels_
print(data['clusters'])
print('\nSum of Square Error/ WCSS is :',Kmeans.inertia_)
```

```python
plt.figure(figsize=(8 ,8))

plt.plot()

plt.scatter(x_axis , y_axis , c=data['clusters'])

plt.title("Cluster ")
plt.grid()

plt.show()
```

```python

print('For Two Cluster, all data points should be assigned to 2 Cluster\n')

kmeans = KMeans(n_clusters = 2, max_iter = 100)
kmeans.fit(data)
data['clusters'] = kmeans.labels_
print('Clusters assigned to each datapoints :')
print(data['clusters'])
print('\nSum of Square Error/ WCSS is :',kmeans.inertia_)
```
